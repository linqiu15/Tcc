```julia
λf(x, y, z) = x^2 + y^2 + z^2 - 2 * x * y - 2 * x * z - 2 * y * z
q0f(E, m1, m2, m3, m4) = abs(m3^2 - m4^2 - m1^2 + m2^2) / (2 * E)
μf(m1, m2) = m1 * m2 / (m1 + m2)

# redefine sqrt so that its cut is along the positive x axis
function xsqrt(x)
    imag(x) >=0 ? sqrt(x+0im) : -sqrt(x-0im)
end
```

# VGT

```julia
qon_0(E)=xsqrt(2*μf(mDsn,mDsn)*(E-mDsn-mDsn))
qon_c(E)=xsqrt(2*μf(mDsc,mDsc)*(E-mDsc-mDsc))
```

```julia
qon_0(mDsn+mDsn-0.01)
```

```julia
xxx,www=gauss(50,0,1);
```

```julia
qi=zeros(ComplexF64,length(xxx)+2);
wi=zeros(ComplexF64,length(xxx));
len=length(xxx);
sublen=length(qi);
```

```julia
function find_qw(E,Λ;rs=1,qi=qi,wi=wi)
    qi[1],qi[2]=qon_0(E)*(-1)^(rs+1),qon_c(E)
    for i in eachindex(xxx)
        qi[i+2]=Λ*xxx[i]
        wi[i]=Λ*www[i]
    end
    nothing
end
```

```julia
Vmat=zeros(ComplexF64,2*sublen,2*sublen);
Gmat=zeros(ComplexF64,2*sublen,2*sublen);
Tmat=zeros(ComplexF64,2*sublen,2*sublen);
```

```julia
function VMat!(E;rs=1,Λ=0.5,J=0, Cv=0.0, Cp=0.0,plist=qi,des=Vmat)
    find_qw(E,Λ;rs=rs)

    for l in eachindex(plist)
        for m in eachindex(plist)
            i,j=0,0
            des[i*sublen+l, j*sublen+m] = potential_DsDsbarn(E,plist[l], plist[m];J=J,Cv=Cv,Cp=Cp)
            i,j=0,1
            des[i*sublen+l, j*sublen+m] = potential_DsDsbarcross(E,plist[l], plist[m];J=J,Cv=Cv,Cp=Cp)
            i,j=1,0
            des[i*sublen+l, j*sublen+m] = potential_DsDsbarcross(E,plist[l], plist[m];J=J,Cv=Cv,Cp=Cp)
            i,j=1,1
            des[i*sublen+l, j*sublen+m] = potential_DsDsbarc(E,plist[l], plist[m];J=J,Cv=Cv,Cp=Cp)
        end
    end
    return nothing
end
```

```julia
Gn(E,q)=1/(2π^2)/(E-mDsn-mDsn-q^2/(2*μf(mDsn,mDsn)))
Gc(E,q)=1/(2π^2)/(E-mDsc-mDsc-q^2/(2*μf(mDsc,mDsc)))
```

```julia
function GMat!(E;rs=1,Λ=0.5,qi=qi,wi=wi,des=Gmat)
    find_qw(E,Λ;rs=rs)
    μ1,μ2=μf(mDsn,mDsn),μf(mDsc,mDsc)
    k1,k2=qi[1],qi[2]
    
    tmp=zero(ComplexF64)
    
    for j in eachindex(wi)
        des[2+j,2+j]=wi[j]*qi[2+j]^2*Gn(E,qi[2+j])
        tmp+=-wi[j]*k1^2*Gn(E,qi[2+j])
    end
    des[1,1]=tmp+μ1*k1/(2π^2)*(log((Λ+k1)/(Λ-k1))-im*π)
    
    tmp=0.0+0.0im
    for j in eachindex(wi)
        des[sublen+2+j,sublen+2+j]=wi[j]*qi[2+j]^2*Gc(E,qi[2+j])
        tmp+=-wi[j]*k2^2*Gc(E,qi[2+j])
    end
    des[2+sublen,2+sublen]=tmp+μ2*k2/(2π^2)*(log((Λ+k2)/(Λ-k2))-im*π)
    
    return nothing
end
```

```julia
function detVG(E;rs=1,Λ=0.5,J=0, Cv=0.0, Cp=0.0)
    VMat!(E;rs=rs,Λ=Λ,J=J,Cv=Cv,Cp=Cp)
    GMat!(E;rs=rs,Λ=Λ)
    return det(LinearAlgebra.I-Vmat*Gmat)
end
```

```julia
function TMat!(E;rs=1,Λ=0.5,J=0, Cv=0.0, Cp=0.0,des=Tmat)
    VMat!(E;rs=rs,Λ=Λ,J=J,Cv=Cv,Cp=Cp)
    GMat!(E;rs=rs,Λ=Λ)
    des .=inv(LinearAlgebra.I-Vmat*Gmat)*Vmat
    nothing
end
```

# Parameter study

```julia
function find_poles(pars)
    _rs,_lam,_cv,_cp,_J=pars
    function f!(F, x)
        e, γ = x
        tmp = detVG(e + im * γ; Λ = _lam,rs = _rs,J=_J,Cv=_cv,Cp=_cp)
        F[1] = real(tmp)
        F[2] = imag(tmp)
    end
    tmp=NLsolve.nlsolve((F, x) -> f!(F, x), [mDsn + mDsn - 1e-3, 0.0])
    return tmp.residual_norm <1e-8 ? (tmp.zero[1]-mDsn-mDsn)+im*tmp.zero[2] : NaN+im*NaN
end
```

```julia
function potential_DsDsbarn(E, p, q; J=0, Cv=0.0, Cp=0.0)
    q01 = q0f(E, mDsn, mDsn, mDsn, mDsn)
    res = zero(ComplexF64)

    res += -gs^2 * pwa_central(p, q, mσ^2 - q01^2)
    res += -β^2 * gV^2 * (1 - q01^2 / mρn^2) / sqrt(16) * pwa_central(p, q, mρn^2 - q01^2) 
#     res+=2 * gλ^2 * gV^2 * pwa_VVV_ss(p, q, mρn^2 - q01^2; a=Cv, J=J)
    res += -β^2 * gV^2 * (1 - q01^2 / mω^2) / sqrt(16) * pwa_central(p, q, mω^2 - q01^2) 
#     res+= 2 * gλ^2 * gV^2 * pwa_VVV_ss(p, q, mω^2 - q01^2; a=Cv, J=J)
    res += -β^2 * gV^2 * (1 - q01^2 / mJψ^2) / sqrt(16) * pwa_central(p, q, mJψ^2 - q01^2) * 2
    res += g^2 / fπ^2 / 2 * pwa_Tensor_ss(p, q, mπn^2 - q01^2; J=J)
    res += g^2 / fπ^2 / 2 * pwa_Tensor_ss(p, q, mη^2 - q01^2; a=Cp, J=J) * 1 / 3
    return res
end

function potential_DsDsbarc(E, p, q; J=0, Cv=0.0, Cp=0.0)
    q01 = q0f(E, mDsc, mDsc, mDsc, mDsc)
    res = zero(ComplexF64)

    res += -gs^2 * pwa_central(p, q, mσ^2 - q01^2)
    res += -β^2 * gV^2 * (1 - q01^2 / mρn^2) / sqrt(16) * pwa_central(p, q, mρn^2 - q01^2) 
#     res+= 2 * gλ^2 * gV^2 * pwa_VVV_ss(p, q, mρn^2 - q01^2; a=Cv, J=J)
    res += -β^2 * gV^2 * (1 - q01^2 / mω^2) / sqrt(16) * pwa_central(p, q, mω^2 - q01^2) 
#     res+= 2 * gλ^2 * gV^2 * pwa_VVV_ss(p, q, mω^2 - q01^2; a=Cv, J=J)
    res += -β^2 * gV^2 * (1 - q01^2 / mJψ^2) / sqrt(16) * pwa_central(p, q, mJψ^2 - q01^2) * 2
    res += g^2 / fπ^2 / 2 * pwa_Tensor_ss(p, q, mπn^2 - q01^2; J=J)
    res += g^2 / fπ^2 / 2 * pwa_Tensor_ss(p, q, mη^2 - q01^2; a=Cp, J=J) * 1 / 3
    return res
end

function potential_DsDsbarcross(E, p, q; J=0, Cv=0.0, Cp=0.0)
    q01 = q0f(E, mDsn, mDsn, mDsc, mDsc)
    res = zero(ComplexF64)

    res += -β^2 * gV^2 * (1 - q01^2 / mρc^2) / sqrt(16) * pwa_central(p, q, mρc^2 - q01^2)*2 
#     res+= 2 * gλ^2 * gV^2 * pwa_VVV_ss(p, q, mρc^2 - q01^2; a=Cv, J=J) * 2
    res += g^2 / fπ^2 * pwa_Tensor_ss(p, q, mπc^2 - q01^2; J=J)
    return res
end
```
