```julia
λf(x, y, z) = x^2 + y^2 + z^2 - 2 * x * y - 2 * x * z - 2 * y * z
q0f(E, m1, m2, m3, m4) = abs(m3^2 - m4^2 - m1^2 + m2^2) / (2 * E)
μf(m1, m2) = m1 * m2 / (m1 + m2)

# redefine sqrt so that its cut is along the positive x axis
function xsqrt(x)
    imag(x) >=0 ? sqrt(x+0im) : -sqrt(x-0im)
end
```

# VGT

```julia
μc,μ0=μf(mDn,mDsc),μf(mDc,mDsn)
```

```julia
qon_c(E)=xsqrt(2*μc*(E-mDn-mDsc+im*ΓDsc/2))
qon_0(E)=xsqrt(2*μ0*(E-mDc-mDsn+im*ΓDsn/2))
```

```julia
qon_c(mDn+mDsc-360e-6-im*24e-6)
```

```julia
xxx,www=gauss(30,0,1);
```

```julia
qi=zeros(ComplexF64,length(xxx)+2);
wi=zeros(ComplexF64,length(xxx));
len=length(xxx);
sublen=length(qi);
```

```julia
function find_qw(E,Λ;rs=1,qi=qi,wi=wi)
    qi[1],qi[2]=qon_c(E)*(-1)^(rs+1),qon_0(E)
    for i in eachindex(xxx)
        qi[i+2]=Λ*xxx[i]
        wi[i]=Λ*www[i]
    end
    nothing
end
```

```julia
find_qw(mDn+mDsc-360e-6-im*24e-6,0.5)
```

```julia
Vmat=zeros(ComplexF64,2*sublen,2*sublen);
Gmat=zeros(ComplexF64,2*sublen,2*sublen);
Tmat=zeros(ComplexF64,2*sublen,2*sublen);
```

```julia
function VMat!(E;rs=1,Λ=0.5,Cv = 0.0,Cp=0.0,I=0,plist=qi,des=Vmat)
    find_qw(E,Λ;rs=rs)
    k1,k2=plist[1],plist[2]

    for l in eachindex(plist)
        for m in eachindex(plist)
            i, j = 1, 1
            des[sublen*(i-1)+l, sublen*(j-1)+m] = potential_DnDsc(E,plist[l], plist[m]; Cv=Cv,Cp=Cp,I=I,isπ=false)
            i, j = 1, 2
            des[sublen*(i-1)+l, sublen*(j-1)+m] = potential_DDscross(E,plist[l], plist[m];  Cv=Cv,Cp=Cp,I=I,isπ=false)
            i, j = 2, 1
            des[sublen*(i-1)+l, sublen*(j-1)+m] = potential_DDscross(E,plist[l], plist[m];  Cv=Cv,Cp=Cp,I=I,isπ=false)
            i, j = 2, 2
            des[sublen*(i-1)+l, sublen*(j-1)+m] = potential_DcDsn(E,plist[l], plist[m];  Cv=Cv,Cp=Cp,I=I,isπ=false)
        end
    end
    return nothing
end
```

```julia
Gc(E,q)=1/(2π^2)/(E-mDn-mDsc-q^2/(2*μf(mDn,mDsc))+im*ΓDsc/2)
Gn(E,q)=1/(2π^2)/(E-mDc-mDsn-q^2/(2*μf(mDc,mDsn))+im*ΓDsn/2)
```

```julia
function GMat!(E;rs=1,Λ=0.5,qi=qi,wi=wi,des=Gmat)
    find_qw(E,Λ;rs=rs)
    μ1,μ2=μf(mDn,mDsc),μf(mDc,mDsn)
    k1,k2=qi[1],qi[2]
    
    tmp=zero(ComplexF64)
    
    i=1
    tmp=0.0+0.0im
    for j in eachindex(wi)
        des[(i-1)*sublen+2+j,(i-1)*sublen+2+j]=wi[j]*qi[2+j]^2*Gc(E,qi[2+j])
        tmp+=-wi[j]*k1^2*Gc(E,qi[2+j])
    end
    des[(i-1)*sublen+1,(i-1)*sublen+1]=tmp+μ1*k1/(2π^2)*(log((Λ+k1)/(Λ-k1))-im*π)
    
    i=2
    tmp=0.0+0.0im
    for j in eachindex(wi)
        des[(i-1)*sublen+2+j,(i-1)*sublen+2+j]=wi[j]*qi[2+j]^2*Gn(E,qi[2+j])
        tmp+=-wi[j]*k2^2*Gn(E,qi[2+j])
    end
    des[(i-1)*sublen+2,(i-1)*sublen+2]=tmp+μ2*k2/(2π^2)*(log((Λ+k2)/(Λ-k2))-im*π)
    
    return nothing
end
```

```julia
function detVG(E;rs=1,Λ=0.5,Cv = 0.0,Cp=0.0,I=0)
    VMat!(E;rs=rs,Λ=Λ,Cv = Cv,Cp=Cp,I=I)
    GMat!(E;rs=rs,Λ=Λ)
    return det(LinearAlgebra.I-Vmat*Gmat)
end
```

```julia
function TMat!(E;rs=1,Λ=0.5,Cv = 0.0,Cp=0.0,I=0,des=Tmat)
    VMat!(E;rs=rs,Λ=Λ,Cv = Cv,Cp=Cp,I=I)
    GMat!(E;rs=rs,Λ=Λ)
    des .=inv(LinearAlgebra.I-Vmat*Gmat)*Vmat
    nothing
end
```

```julia
using Optim;
```

```julia
function inv_abst(x;rs=1,E=mDn+mDsc-273e-6,Cv=0.0,Cp=0.0,I=0)
    TMat!(E;rs=rs,Λ=x[1],Cv = Cv,Cp=Cp,I=I)
    return 1/abs(Tmat[1,1])
end
```

```julia
function absdet(x;rs=1,E=mDn+mDsc-360e-6-im*24e-6,Cv=0.0,Cp=0.0,I=0)
    abs(detVG(E;rs=rs,Λ=x[1],Cv = Cv,Cp=Cp,I=I))
end
```

```julia
function find_poles(pars)
    _rs,_lam,Cv,Cp,I=pars
    function f!(F, x)
        e, γ = x
        tmp = detVG(e + im * γ; Λ = _lam,rs = _rs,Cv = Cv,Cp=Cp,I=I)
        F[1] = real(tmp)
        F[2] = imag(tmp)
    end
    tmp=NLsolve.nlsolve((F, x) -> f!(F, x), [mDn + mDsc - 333e-6, 0.0])
#     (tmp.zero[1]-mDn-mDsc)+im*tmp.zero[2]
end
```

```julia
let
    tmpf(x;Cv=0.0,Cp=0.0)=inv_abst(x;Cv=Cv,Cp=Cp)
    function findΛ(Cv,Cp)
        res=Optim.optimize(λ->tmpf(λ;Cv=Cv,Cp=Cp),[0.3],[1.5],[0.5])
        return Optim.minimizer(res)[1]
    end
    
    vec_Cv=collect(0.0:0.1:1.0)
    vec_Cp=collect(0.0:0.1:1.0)
    vec_res=[findΛ(x,y) for y in vec_Cp,x in vec_Cv]
    surface(vec_Cv,vec_Cp,vec_res)
end
```
